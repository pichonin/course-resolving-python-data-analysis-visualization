## [ ] Load the data

```python
import pandas as pd

df_premier = pd.read_excel('premier_league.xlsx', index_col=0)
df_premier
```

$$
points = (a) + (b) \cdot goals
$$

## Select the variables for the model

1. y: the variable you want to predict
2. X: the variable you will use to calculate the prediction y

```python
y = df_premier.points
```

```python
X = df_premier[['goals']]
```

## The Linear Regression ML Model

### Fit the model with the data

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(X, y)
```

### Predictions

#### Calculate the predictions

```python
model_lr.predict(X)
```

### Model evaluation

#### Calculate the model's score

```python
model_lr.score(X, y)
```

## The K Nearest Neighbours ML Model

### Fit the model with the data

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model_kn = KNeighborsRegressor()
```

```python
model_kn.fit(X, y)
```

### Predictions

#### Calculate the predictions

```python
model_kn.predict(X)
```

### Model evaluation

#### Calculate the model's score

```python
model_kn.score(X, y)
```

## The Random Forest ML Model

### Fit the model with the data

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_rf = RandomForestRegressor()
```

```python
model_rf.fit(X, y)
```

### Predictions

#### Calculate the predictions

```python
model_rf.predict(X)
```

### Model evaluation

#### Calculate the model's score

```python
model_rf.score(X, y)
```
