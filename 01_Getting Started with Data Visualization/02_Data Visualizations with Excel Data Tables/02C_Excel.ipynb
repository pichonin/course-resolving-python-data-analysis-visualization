## Create a scatter plot

### Import matplotlib library

```python
import matplotlib.pyplot as plt
```

### Access column data separately

```python
df_gap.money_per_person
```

```python
df_gap.life_expectancy
```

### Pass data to scatter function

```python
plt.scatter(x= df_gap.money_per_person , y= df_gap.life_expectancy)
```

## Customize the plot

### Change the size

```python
plt.scatter(x= df_gap.money_per_person , y= df_gap.life_expectancy , s= 100)
```

### Change the color

```python
plt.scatter(x= df_gap.money_per_person , y= df_gap.life_expectancy , s=100 , c='red')
```

### Change the marker

```python
plt.scatter(x= df_gap.money_per_person , y= df_gap.life_expectancy , s=100 , c='red' , marker='x')
```

### Change the border

```python
plt.scatter(x= df_gap.money_per_person , y= df_gap.life_expectancy , s=200 , c='red' , marker='X' , edgecolors='black' , )
```
